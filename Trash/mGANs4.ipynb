```python
# MGAN Implementation in PyTorch for MNIST Dataset

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms

import matplotlib.pyplot as plt
import numpy as np

from torch.utils.data import DataLoader
import torchvision.utils as vutils
import os

# Set random seed for reproducibility
manualSeed = 6789
torch.manual_seed(manualSeed)
np.random.seed(manualSeed)

# Define the Generator class
class Generator(nn.Module):
    def __init__(self, nz, ngf, nc, num_conv_layers):
        super(Generator, self).__init__()
        self.nz = nz
        self.ngf = ngf
        self.nc = nc
        self.num_conv_layers = num_conv_layers

        # Input layer is unique per generator
        self.input_layer = nn.Sequential(
            nn.Linear(self.nz, self.ngf * 4 * 4),
            nn.BatchNorm1d(self.ngf * 4 * 4),
            nn.ReLU(True)
        )

        # Shared layers
        self.shared_layers = nn.Sequential(
            nn.ConvTranspose2d(self.ngf, self.ngf//2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ngf//2),
            nn.ReLU(True),
            nn.ConvTranspose2d(self.ngf//2, self.ngf//4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ngf//4),
            nn.ReLU(True),
            nn.ConvTranspose2d(self.ngf//4, self.nc, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, input):
        x = self.input_layer(input)
        x = x.view(-1, self.ngf, 4, 4)
        x = self.shared_layers(x)
        return x

    def share_weights_from(self, other_gen):
        # Share weights except input layer
        self.shared_layers.load_state_dict(other_gen.shared_layers.state_dict())

# Define the Discriminator class
class Discriminator(nn.Module):
    def __init__(self, nc, ndf, num_conv_layers, num_gens):
        super(Discriminator, self).__init__()
        self.nc = nc
        self.ndf = ndf
        self.num_conv_layers = num_conv_layers
        self.num_gens = num_gens

        # Shared layers
        self.shared_layers = nn.Sequential(
            nn.Conv2d(self.nc, self.ndf//4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf//4),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(self.ndf//4, self.ndf//2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf//2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(self.ndf//2, self.ndf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf),
            nn.LeakyReLU(0.2, inplace=True),
        )

        # Output layers
        self.output_bin = nn.Linear(self.ndf * 4 * 4, 1)
        self.output_mul = nn.Linear(self.ndf * 4 * 4, self.num_gens)

    def forward(self, input):
        x = self.shared_layers(input)
        x = x.view(-1, self.ndf * 4 * 4)
        output_bin = self.output_bin(x)
        output_mul = self.output_mul(x)
        return output_bin, output_mul

# Define the MGAN class
class MGAN:
    """Mixture Generative Adversarial Nets"""
    def __init__(self,
                 num_z=128,
                 beta=1.0,
                 num_gens=4,
                 d_batch_size=64,
                 g_batch_size=32,
                 z_prior="uniform",
                 learning_rate=0.0002,
                 img_size=(32, 32, 1),
                 num_conv_layers=3,
                 num_gen_feature_maps=128,
                 num_dis_feature_maps=128,
                 num_epochs=500,
                 sample_dir="samples",
                 sample_by_gen_dir="samples_by_gen",
                 random_seed=6789,
                 device='cpu'):
        self.beta = beta
        self.num_z = num_z
        self.num_gens = num_gens
        self.d_batch_size = d_batch_size
        self.g_batch_size = g_batch_size
        self.z_prior = z_prior
        self.learning_rate = learning_rate
        self.num_epochs = num_epochs
        self.img_size = img_size
        self.num_conv_layers = num_conv_layers
        self.num_gen_feature_maps = num_gen_feature_maps
        self.num_dis_feature_maps = num_dis_feature_maps
        self.sample_dir = sample_dir
        self.sample_by_gen_dir = sample_by_gen_dir
        self.random_seed = random_seed
        self.device = device

        self.history = {'d_loss': [], 'd_bin_loss': [], 'd_mul_loss': [],
                        'g_loss': [], 'g_bin_loss': [], 'g_mul_loss': []}

        self._build_model()

    def _build_model(self):
        # Build Generators
        self.generators = nn.ModuleList([Generator(self.num_z, self.num_gen_feature_maps,
                                                   self.img_size[2], self.num_conv_layers).to(self.device)
                                         for _ in range(self.num_gens)])

        # Share parameters among generators except input layers
        for i in range(1, self.num_gens):
            self.generators[i].share_weights_from(self.generators[0])

        # Build Discriminator and Classifier
        self.discriminator = Discriminator(self.img_size[2], self.num_dis_feature_maps,
                                           self.num_conv_layers, self.num_gens).to(self.device)

        # Optimizers
        self.optimizerD = optim.Adam(self.discriminator.parameters(), lr=self.learning_rate, betas=(0.5, 0.999))

        self.optimizerGs = []
        for gen in self.generators:
            optimizerG = optim.Adam(gen.parameters(), lr=self.learning_rate, betas=(0.5, 0.999))
            self.optimizerGs.append(optimizerG)

        # Loss functions
        self.criterion_bin = nn.BCEWithLogitsLoss()
        self.criterion_mul = nn.CrossEntropyLoss()

    def fit(self, trainloader):
        fixed_noise = self._sample_z(self.g_batch_size * self.num_gens).to(self.device)

        for epoch in range(self.num_epochs):
            for i, data in enumerate(trainloader, 0):
                ############################
                # (1) Update D network
                ###########################
                self.discriminator.zero_grad()
                real_images = data[0].to(self.device)
                b_size = real_images.size(0)

                # Create real labels as floats
                label_real = torch.full((b_size,), 1.0, device=self.device, dtype=torch.float)

                # Forward pass real batch through D
                output_bin_real, _ = self.discriminator(real_images)
                d_bin_x_loss = self.criterion_bin(output_bin_real.squeeze(), label_real)

                # Generate fake images
                fake_images = []
                gen_labels = []
                for idx, gen in enumerate(self.generators):
                    z = self._sample_z(self.g_batch_size).to(self.device)
                    gen_imgs = gen(z)
                    fake_images.append(gen_imgs)
                    gen_labels.append(torch.full((self.g_batch_size,), idx, dtype=torch.long, device=self.device))

                fake_images = torch.cat(fake_images, 0)
                gen_labels = torch.cat(gen_labels, 0)

                # Create fake labels as floats
                label_fake = torch.full((fake_images.size(0),), 0.0, device=self.device, dtype=torch.float)

                # Forward pass fake batch through D
                output_bin_fake, output_mul_fake = self.discriminator(fake_images.detach())
                d_bin_g_loss = self.criterion_bin(output_bin_fake.squeeze(), label_fake)
                d_mul_loss = self.criterion_mul(output_mul_fake, gen_labels)

                # Sum all discriminator losses
                d_loss = d_bin_x_loss + d_bin_g_loss + d_mul_loss
                d_loss.backward()

                self.optimizerD.step()

                ############################
                # (2) Update G network
                ###########################
                # Update each generator individually
                for idx, gen in enumerate(self.generators):
                    gen.zero_grad()
                    # We want the generator to fool the discriminator
                    label_gen = torch.full((self.g_batch_size,), 1.0, device=self.device, dtype=torch.float)
                    z = self._sample_z(self.g_batch_size).to(self.device)
                    gen_imgs = gen(z)
                    output_bin_fake, output_mul_fake = self.discriminator(gen_imgs)
                    g_bin_loss = self.criterion_bin(output_bin_fake.squeeze(), label_gen)
                    gen_labels = torch.full((self.g_batch_size,), idx, dtype=torch.long, device=self.device)
                    g_mul_loss = self.beta * self.criterion_mul(output_mul_fake, gen_labels)
                    g_loss = g_bin_loss + g_mul_loss
                    g_loss.backward()
                    self.optimizerGs[idx].step()

                    # Save generator losses
                    self.history['g_bin_loss'].append(g_bin_loss.item())
                    self.history['g_mul_loss'].append(g_mul_loss.item())
                    self.history['g_loss'].append(g_loss.item())

                # Save discriminator losses
                self.history['d_bin_loss'].append(d_bin_x_loss.item() + d_bin_g_loss.item())
                self.history['d_mul_loss'].append(d_mul_loss.item())
                self.history['d_loss'].append(d_loss.item())

            # Output training stats
            print('[%d/%d] d_loss: %.4f | d_bin_loss: %.4f | d_mul_loss: %.4f | '
                  'g_loss: %.4f | g_bin_loss: %.4f | g_mul_loss: %.4f'
                  % (epoch+1, self.num_epochs,
                     self.history['d_loss'][-1],
                     self.history['d_bin_loss'][-1],
                     self.history['d_mul_loss'][-1],
                     self.history['g_loss'][-1],
                     self.history['g_bin_loss'][-1],
                     self.history['g_mul_loss'][-1]))

            # Save samples every 5 epochs
            if (epoch+1) % 5 == 0:
                self._save_samples(epoch+1, fixed_noise)

        # After training, plot the learning curves
        self._plot_history()

    def _sample_z(self, size):
        if self.z_prior == "uniform":
            return torch.rand(size, self.num_z) * 2 - 1  # Uniform between [-1, 1]
        else:
            return torch.randn(size, self.num_z)

    def _save_samples(self, epoch, fixed_noise):
        # Create directories if they don't exist
        if not os.path.exists(self.sample_dir):
            os.makedirs(self.sample_dir)
        if not os.path.exists(self.sample_by_gen_dir):
            os.makedirs(self.sample_by_gen_dir)

        # Generate images
        with torch.no_grad():
            fake_images_list = []
            for gen in self.generators:
                fake_images = gen(fixed_noise[:self.g_batch_size].to(self.device))
                fake_images_list.append(fake_images)

            # Concatenate images from all generators
            fake_images = torch.cat(fake_images_list, 0)

            # Normalize images to range [0, 1]
            fake_images = (fake_images + 1) / 2.0

            # Save the images
            sample_path = os.path.join(self.sample_dir, 'epoch_{:04d}.png'.format(epoch))
            vutils.save_image(fake_images, sample_path, nrow=self.num_gens, padding=2, normalize=True)

            # Save images by generator
            images_by_gen = []
            for fake_images in fake_images_list:
                images_by_gen.append(fake_images)
            images_by_gen = torch.cat(images_by_gen, 0)
            sample_by_gen_path = os.path.join(self.sample_by_gen_dir, 'epoch_{:04d}.png'.format(epoch))
            vutils.save_image(images_by_gen, sample_by_gen_path, nrow=self.g_batch_size, padding=2, normalize=True)

    def _plot_history(self):
        # Plot the learning curves
        epochs = range(1, self.num_epochs+1)

        # Plot Discriminator Loss
        plt.figure(figsize=(10,5))
        plt.title("Discriminator Loss During Training")
        plt.plot(epochs, self.history['d_loss'], label="D Loss")
        plt.plot(epochs, self.history['d_bin_loss'], label="D Binary Loss")
        plt.plot(epochs, self.history['d_mul_loss'], label="D Multi-class Loss")
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

        # Plot Generator Loss
        plt.figure(figsize=(10,5))
        plt.title("Generator Loss During Training")
        plt.plot(epochs, self.history['g_loss'], label="G Loss")
        plt.plot(epochs, self.history['g_bin_loss'], label="G Binary Loss")
        plt.plot(epochs, self.history['g_mul_loss'], label="G Multi-class Loss")
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

# Define the main function
def main():
    # Hyperparameters
    num_z = 100  # Size of z latent vector (i.e. size of generator input)
    beta = 0.01  # Diversity parameter beta
    num_gens = 10  # Number of generators
    d_batch_size = 64  # Batch size for discriminator
    g_batch_size = 12  # Batch size for generators
    z_prior = "uniform"  # Prior distribution of the noise ('uniform' or 'gaussian')
    learning_rate = 0.0002  # Learning rate for optimizers
    num_conv_layers = 3  # Number of convolutional layers
    num_gen_feature_maps = 128  # Number of feature maps in generator
    num_dis_feature_maps = 128  # Number of feature maps in discriminator
    num_epochs = 500  # Number of training epochs
    image_size = 32  # Spatial size of training images
    num_channels = 1  # Number of channels in the training images. For MNIST, it's 1

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print("Using device:", device)

    # Create the dataset
    dataroot = './data'
    transform = transforms.Compose([
        transforms.Resize(image_size),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))  # Normalize the images to [-1, 1]
    ])

    # Load the dataset
    trainset = torchvision.datasets.MNIST(root=dataroot, train=True,
                                          download=True, transform=transform)
    trainloader = DataLoader(trainset, batch_size=d_batch_size,
                             shuffle=True, num_workers=2)

    # Create the MGAN model
    mgan_model = MGAN(
        num_z=num_z,
        beta=beta,
        num_gens=num_gens,
        d_batch_size=d_batch_size,
        g_batch_size=g_batch_size,
        z_prior=z_prior,
        learning_rate=learning_rate,
        img_size=(image_size, image_size, num_channels),
        num_conv_layers=num_conv_layers,
        num_gen_feature_maps=num_gen_feature_maps,
        num_dis_feature_maps=num_dis_feature_maps,
        num_epochs=num_epochs,
        sample_dir="samples",
        sample_by_gen_dir="samples_by_gen",
        random_seed=manualSeed,
        device=device
    )

    # Train the model
    mgan_model.fit(trainloader)
# Import necessary libraries
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms

import matplotlib.pyplot as plt
import numpy as np

from torch.utils.data import DataLoader
import torchvision.utils as vutils
import os

# Set random seed for reproducibility
manualSeed = 6789
torch.manual_seed(manualSeed)
np.random.seed(manualSeed)

# Define the Generator class
class Generator(nn.Module):
    def __init__(self, nz, ngf, nc, num_conv_layers):
        super(Generator, self).__init__()
        self.nz = nz
        self.ngf = ngf
        self.nc = nc
        self.num_conv_layers = num_conv_layers

        # Input layer is unique per generator
        self.input_layer = nn.Sequential(
            nn.Linear(self.nz, self.ngf * 4 * 4),
            nn.BatchNorm1d(self.ngf * 4 * 4),
            nn.ReLU(True)
        )

        # Shared layers
        self.shared_layers = nn.Sequential(
            nn.ConvTranspose2d(self.ngf, self.ngf//2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ngf//2),
            nn.ReLU(True),
            nn.ConvTranspose2d(self.ngf//2, self.ngf//4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ngf//4),
            nn.ReLU(True),
            nn.ConvTranspose2d(self.ngf//4, self.nc, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, input):
        x = self.input_layer(input)
        x = x.view(-1, self.ngf, 4, 4)
        x = self.shared_layers(x)
        return x

    def share_weights_from(self, other_gen):
        # Share weights except input layer
        self.shared_layers.load_state_dict(other_gen.shared_layers.state_dict())

# Define the Discriminator class
class Discriminator(nn.Module):
    def __init__(self, nc, ndf, num_conv_layers, num_gens):
        super(Discriminator, self).__init__()
        self.nc = nc
        self.ndf = ndf
        self.num_conv_layers = num_conv_layers
        self.num_gens = num_gens

        # Shared layers
        self.shared_layers = nn.Sequential(
            nn.Conv2d(self.nc, self.ndf//4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf//4),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(self.ndf//4, self.ndf//2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf//2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(self.ndf//2, self.ndf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf),
            nn.LeakyReLU(0.2, inplace=True),
        )

        # Output layers
        self.output_bin = nn.Linear(self.ndf * 4 * 4, 1)
        self.output_mul = nn.Linear(self.ndf * 4 * 4, self.num_gens)

    def forward(self, input):
        x = self.shared_layers(input)
        x = x.view(-1, self.ndf * 4 * 4)
        output_bin = self.output_bin(x)
        output_mul = self.output_mul(x)
        return output_bin, output_mul

# Define the MGAN class
class MGAN:
    """Mixture Generative Adversarial Nets"""
    def __init__(self,
                 num_z=128,
                 beta=1.0,
                 num_gens=4,
                 d_batch_size=64,
                 g_batch_size=32,
                 z_prior="uniform",
                 learning_rate=0.0002,
                 img_size=(32, 32, 1),
                 num_conv_layers=3,
                 num_gen_feature_maps=128,
                 num_dis_feature_maps=128,
                 num_epochs=500,
                 sample_dir="samples",
                 sample_by_gen_dir="samples_by_gen",
                 random_seed=6789,
                 device='cpu'):
        self.beta = beta
        self.num_z = num_z
        self.num_gens = num_gens
        self.d_batch_size = d_batch_size
        self.g_batch_size = g_batch_size
        self.z_prior = z_prior
        self.learning_rate = learning_rate
        self.num_epochs = num_epochs
        self.img_size = img_size
        self.num_conv_layers = num_conv_layers
        self.num_gen_feature_maps = num_gen_feature_maps
        self.num_dis_feature_maps = num_dis_feature_maps
        self.sample_dir = sample_dir
        self.sample_by_gen_dir = sample_by_gen_dir
        self.random_seed = random_seed
        self.device = device

        self.history = {'d_loss': [], 'd_bin_loss': [], 'd_mul_loss': [],
                        'g_loss': [], 'g_bin_loss': [], 'g_mul_loss': []}

        self._build_model()

    def _build_model(self):
        # Build Generators
        self.generators = nn.ModuleList([Generator(self.num_z, self.num_gen_feature_maps,
                                                   self.img_size[2], self.num_conv_layers).to(self.device)
                                         for _ in range(self.num_gens)])

        # Share parameters among generators except input layers
        for i in range(1, self.num_gens):
            self.generators[i].share_weights_from(self.generators[0])

        # Build Discriminator and Classifier
        self.discriminator = Discriminator(self.img_size[2], self.num_dis_feature_maps,
                                           self.num_conv_layers, self.num_gens).to(self.device)

        # Optimizers
        self.optimizerD = optim.Adam(self.discriminator.parameters(), lr=self.learning_rate, betas=(0.5, 0.999))

        self.optimizerGs = []
        for gen in self.generators:
            optimizerG = optim.Adam(gen.parameters(), lr=self.learning_rate, betas=(0.5, 0.999))
            self.optimizerGs.append(optimizerG)

        # Loss functions
        self.criterion_bin = nn.BCEWithLogitsLoss()
        self.criterion_mul = nn.CrossEntropyLoss()

    def fit(self, trainloader):
        fixed_noise = self._sample_z(self.g_batch_size * self.num_gens).to(self.device)

        for epoch in range(self.num_epochs):
            for i, data in enumerate(trainloader, 0):
                ############################
                # (1) Update D network
                ###########################
                self.discriminator.zero_grad()
                real_images = data[0].to(self.device)
                b_size = real_images.size(0)

                # Create real labels as floats
                label_real = torch.full((b_size,), 1.0, device=self.device, dtype=torch.float)

                # Forward pass real batch through D
                output_bin_real, _ = self.discriminator(real_images)
                d_bin_x_loss = self.criterion_bin(output_bin_real.squeeze(), label_real)

                # Generate fake images
                fake_images = []
                gen_labels = []
                for idx, gen in enumerate(self.generators):
                    z = self._sample_z(self.g_batch_size).to(self.device)
                    gen_imgs = gen(z)
                    fake_images.append(gen_imgs)
                    gen_labels.append(torch.full((self.g_batch_size,), idx, dtype=torch.long, device=self.device))

                fake_images = torch.cat(fake_images, 0)
                gen_labels = torch.cat(gen_labels, 0)

                # Create fake labels as floats
                label_fake = torch.full((fake_images.size(0),), 0.0, device=self.device, dtype=torch.float)

                # Forward pass fake batch through D
                output_bin_fake, output_mul_fake = self.discriminator(fake_images.detach())
                d_bin_g_loss = self.criterion_bin(output_bin_fake.squeeze(), label_fake)
                d_mul_loss = self.criterion_mul(output_mul_fake, gen_labels)

                # Sum all discriminator losses
                d_loss = d_bin_x_loss + d_bin_g_loss + d_mul_loss
                d_loss.backward()

                self.optimizerD.step()

                ############################
                # (2) Update G network
                ###########################
                # Update each generator individually
                for idx, gen in enumerate(self.generators):
                    gen.zero_grad()
                    # We want the generator to fool the discriminator
                    label_gen = torch.full((self.g_batch_size,), 1.0, device=self.device, dtype=torch.float)
                    z = self._sample_z(self.g_batch_size).to(self.device)
                    gen_imgs = gen(z)
                    output_bin_fake, output_mul_fake = self.discriminator(gen_imgs)
                    g_bin_loss = self.criterion_bin(output_bin_fake.squeeze(), label_gen)
                    gen_labels = torch.full((self.g_batch_size,), idx, dtype=torch.long, device=self.device)
                    g_mul_loss = self.beta * self.criterion_mul(output_mul_fake, gen_labels)
                    g_loss = g_bin_loss + g_mul_loss
                    g_loss.backward()
                    self.optimizerGs[idx].step()

                    # Save generator losses
                    self.history['g_bin_loss'].append(g_bin_loss.item())
                    self.history['g_mul_loss'].append(g_mul_loss.item())
                    self.history['g_loss'].append(g_loss.item())

                # Save discriminator losses
                self.history['d_bin_loss'].append(d_bin_x_loss.item() + d_bin_g_loss.item())
                self.history['d_mul_loss'].append(d_mul_loss.item())
                self.history['d_loss'].append(d_loss.item())

            # Output training stats
            print('[%d/%d] d_loss: %.4f | d_bin_loss: %.4f | d_mul_loss: %.4f | '
                  'g_loss: %.4f | g_bin_loss: %.4f | g_mul_loss: %.4f'
                  % (epoch+1, self.num_epochs,
                     self.history['d_loss'][-1],
                     self.history['d_bin_loss'][-1],
                     self.history['d_mul_loss'][-1],
                     self.history['g_loss'][-1],
                     self.history['g_bin_loss'][-1],
                     self.history['g_mul_loss'][-1]))

            # Save and visualize samples every 5 epochs
            if (epoch+1) % 5 == 0:
                self._save_and_plot_samples(epoch+1, fixed_noise)

        # After training, plot the learning curves
        self._plot_history()

    def _sample_z(self, size):
        if self.z_prior == "uniform":
            return torch.rand(size, self.num_z) * 2 - 1  # Uniform between [-1, 1]
        else:
            return torch.randn(size, self.num_z)

    def _save_and_plot_samples(self, epoch, fixed_noise):
        # Create directories if they don't exist
        if not os.path.exists(self.sample_dir):
            os.makedirs(self.sample_dir)
        if not os.path.exists(self.sample_by_gen_dir):
            os.makedirs(self.sample_by_gen_dir)

        # Generate images
        with torch.no_grad():
            fake_images_list = []
            for gen in self.generators:
                fake_images = gen(fixed_noise[:self.g_batch_size].to(self.device))
                fake_images_list.append(fake_images)

            # Concatenate images from all generators
            fake_images = torch.cat(fake_images_list, 0)

            # Normalize images to range [0, 1]
            fake_images = (fake_images + 1) / 2.0

            # Save the images
            sample_path = os.path.join(self.sample_dir, 'epoch_{:04d}.png'.format(epoch))
            vutils.save_image(fake_images, sample_path, nrow=self.num_gens, padding=2, normalize=True)

            # Plot images by generator
            images_by_gen = []
            for fake_images in fake_images_list:
                images_by_gen.append(fake_images)
            images_by_gen = torch.cat(images_by_gen, 0)
            sample_by_gen_path = os.path.join(self.sample_by_gen_dir, 'epoch_{:04d}.png'.format(epoch))
            vutils.save_image(images_by_gen, sample_by_gen_path, nrow=self.g_batch_size, padding=2, normalize=True)

            # Visualize generated images inline
            plt.figure(figsize=(15, 15))
            grid = vutils.make_grid(fake_images, nrow=self.num_gens, padding=2, normalize=True)
            plt.imshow(np.transpose(grid.cpu(), (1, 2, 0)))
            plt.title(f'Generated Images at Epoch {epoch}')
            plt.axis('off')
            plt.show()

    def _plot_history(self):
        # Plot the learning curves
        epochs = range(1, self.num_epochs+1)

        # Plot Discriminator Loss
        plt.figure(figsize=(10,5))
        plt.title("Discriminator Loss During Training")
        plt.plot(epochs, self.history['d_loss'], label="D Loss")
        plt.plot(epochs, self.history['d_bin_loss'], label="D Binary Loss")
        plt.plot(epochs, self.history['d_mul_loss'], label="D Multi-class Loss")
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

        # Plot Generator Loss
        plt.figure(figsize=(10,5))
        plt.title("Generator Loss During Training")
        plt.plot(epochs, self.history['g_loss'], label="G Loss")
        plt.plot(epochs, self.history['g_bin_loss'], label="G Binary Loss")
        plt.plot(epochs, self.history['g_mul_loss'], label="G Multi-class Loss")
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

# Define the main function
def main():
    # Hyperparameters
    num_z = 100  # Size of z latent vector (i.e. size of generator input)
    beta = 0.01  # Diversity parameter beta
    num_gens = 10  # Number of generators
    d_batch_size = 64  # Batch size for discriminator
    g_batch_size = 12  # Batch size for generators
    z_prior = "uniform"  # Prior distribution of the noise ('uniform' or 'gaussian')
    learning_rate = 0.0002  # Learning rate for optimizers
    num_conv_layers = 3  # Number of convolutional layers
    num_gen_feature_maps = 128  # Number of feature maps in generator
    num_dis_feature_maps = 128  # Number of feature maps in discriminator
    num_epochs = 500  # Number of training epochs
    image_size = 32  # Spatial size of training images
    num_channels = 1  # Number of channels in the training images. For MNIST, it's 1

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print("Using device:", device)

    # Create the dataset
    dataroot = './data'
    transform = transforms.Compose([
        transforms.Resize(image_size),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))  # Normalize the images to [-1, 1]
    ])

    # Load the dataset
    trainset = torchvision.datasets.MNIST(root=dataroot, train=True,
                                          download=True, transform=transform)
    trainloader = DataLoader(trainset, batch_size=d_batch_size,
                             shuffle=True, num_workers=2)

    # Create the MGAN model
    mgan_model = MGAN(
        num_z=num_z,
        beta=beta,
        num_gens=num_gens,
        d_batch_size=d_batch_size,
        g_batch_size=g_batch_size,
        z_prior=z_prior,
        learning_rate=learning_rate,
        img_size=(image_size, image_size, num_channels),
        num_conv_layers=num_conv_layers,
        num_gen_feature_maps=num_gen_feature_maps,
        num_dis_feature_maps=num_dis_feature_maps,
        num_epochs=num_epochs,
        sample_dir="samples",
        sample_by_gen_dir="samples_by_gen",
        random_seed=manualSeed,
        device=device
    )

    # Train the model
    mgan_model.fit(trainloader)

# Execute the main function
if __name__ == '__main__':
    main()

if __name__ == '__main__':
    main()
```
